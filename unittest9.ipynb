```python
# q 2) The average IQ is 100 with a standard deviation of 15, using the emerical rule , what percent have IQ greater than 130?

# Ans:

from scipy.stats import norm
mean=130
std_dev=15

# probability of IQ greater than 130
probability =1-norm.cdf(130,mean,std_dev)
print("percentage of people wit IQ greater than 130:",probability*100)
```
